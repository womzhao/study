```python
import pandas as pd
import plotly.express as px
```

```python
df0 = pd.read_excel('Sales-tables-December-2022-quarter.xlsx', sheet_name='Postcode',skiprows=6)   
postcode = pd.read_csv('au_postcodes.csv')
```

```python
postcode = postcode.groupby('postcode').agg({'place_name':'first', 'state_name': 'first', 'state_code':'first','latitude': 'mean', 'longitude':'mean'})
```

```python
columns_name = ['postcode', 'dwelling', 'first quartile(000)', 'median', 'third quartile', 'mean', '#sales', 'qty chang in median', 'annual change in median', 'qty change in count', 'annual change in count']
df0.columns = columns_name
```

```python
df1 = df0[df0['dwelling'] == 'Non Strata']
```

```python
df1['median'] = df1['median'].replace('-', float('nan'))
df1['median'] = df1['median'].astype(float)
```

```python
df2 = df1.merge(postcode, how='left', on='postcode')
```

```python
fig = px.scatter(df1, x='postcode', y='median', color = 'postcode', range_y=[df1['median'].min(), df1['median'].max()])
fig.show()
```

px.scatter(df1, x='postcode', y='')

```python
df3 = df2[(df2['median'] >= 0) & (df2['median'] < 6000)]
```

```python
# Create the geographic map
fig = px.scatter_mapbox(
    df3, 
    lat='latitude',  # Name of the column in your dataset that contains latitude values
    lon='longitude',  # Name of the column in your dataset that contains longitude values
    hover_name='place_name',  # Name of the column in your dataset to display as hover labels
    hover_data=['median'],  # Additional columns to display in the hover tooltip (optional)
    color_discrete_map=True,
    color='median',  # Name of the column in your dataset to use for color encoding
    zoom=5  # Zoom level of the initial map view
)

# Update the layout of the map
fig.update_layout(
    mapbox_style='carto-positron',  # Mapbox style, you can choose different styles
    margin={'r': 0, 't': 30, 'l': 0, 'b': 0}  # Margin around the map
)

# Show the map
fig.show()
```
